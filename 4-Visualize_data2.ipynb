```python
# -*- coding: utf-8 -*-
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Задача для конкурсного потока

Загрузим данные

```python
path_to_train = '../../data/task_2/gps_train.tsv'
path_to_test = '../../data/task_2/gps_test.tsv'
path_to_stops = '../../data/task_2/stops_train.tsv'

train_df = pd.read_csv(path_to_train, sep='\t', names=['TIME', 'ID', 'TYPE', 'ROUTE', 'X', 'Y'])
test_df = pd.read_csv(path_to_test, sep='\t', names=['TIME', 'ID', 'TYPE', 'ROUTE', 'X', 'Y'])
stops_df = pd.read_csv(path_to_stops, sep='\t', names = ['X', 'Y'])

print train_df.shape, test_df.shape, stops_df.shape
```

Изобразим train gps координаты и координаты остановок

```python
plt.figure(figsize = (16, 8))

plt.subplot(1, 2, 1)
scatter_train = plt.scatter(train_df.X, train_df.Y, s=15)
plt.legend([scatter_train], ['train gps'])

plt.subplot(1, 2, 2)
scatter_train = plt.scatter(train_df.X, train_df.Y, s=15)
scatter_stops = plt.scatter(stops_df.X, stops_df.Y, s=10, color='red')
plt.legend([scatter_train, scatter_stops], ['train gps', 'train stops'])
```

Train + test + train_stops

```python
plt.figure(figsize = (10, 10))
scatter_train = plt.scatter(train_df.X, train_df.Y, s=15)
scatter_test = plt.scatter(test_df.X, test_df.Y, s=15, color='green')
scatter_stops = plt.scatter(stops_df.X, stops_df.Y, s=10, color='red')
plt.legend([scatter_train, scatter_stops, scatter_test], ['train gps', 'train stops', 'test gps'])
```
